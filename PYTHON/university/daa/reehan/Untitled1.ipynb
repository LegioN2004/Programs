```python
def largest(arr, n):
    max = arr[0]
    for i in range(n):
        if arr[i]>max:
            max = arr[i]
    return max
print(largest(arr, len(arr)))
```

```python
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
B = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]
C = []

for i in range(len(A)):
    row = []  # new row for each i
    for j in range(len(A[i])):
        row.append(A[i][j] + B[i][j])
    C.append(row)

print(C)
```

```python
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
B = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

C = []   # result matrix

for i in range(len(A)):           # iterate rows of A
    row = []                      # start a new row
    for j in range(len(B[0])):    # iterate columns of B
        result = 0
        for k in range(len(A[0])):  # dot product
            result += A[i][k] * B[k][j]
        row.append(result)        # add element into the row
    C.append(row)                 # add completed row into C

print(C)
```

```python
numbers = [11, 12, 22, 5,30, 45, 3, 19, 23]
prime = []
for i in numbers:
    is_prime = True
    for j in range(2, i-1):
        if i%j == 0:
            is_prime = False
            break
    if is_prime:
        prime.append(i)
print(f"Prime = {prime}")
```
