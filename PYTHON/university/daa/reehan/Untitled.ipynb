```python
def factorial(num: int):
    if num == 0:
        return 1
    else:
        return num*factorial(num-1)

result = factorial(5)
print(result)
```

```python
def sum(num: int):
    if num == 1:
        return 1
    else:
        return num*num + sum(num-1)
a = int(input("enter a number:"))
result = sum(a)
print(result)
```

```python
# Python3 Program to find the sum
# of sum of squares of first n 
# natural number

# Function to find sum of sum of
# square of first n natural number
def findSum(n):
    summ = 0
    for i in range(1, n+1):
        summ = (summ + ((i * (i + 1) 
                * (2 * i + 1)) / 6)) 
    return summ 

# Driven Program
n = 3
print(int(findSum(n)))


# This code is contributed by 
# Prasad Kshirsagar
```

```python
sum = 0
num = int(input("enter a number:"))
for i in range(num+1):
    sum = sum + i*i*i
print(f"Sum of cube of {num} natural numbers = {sum}")
```

```python
def sum(num: int):
    if num == 1:
        return 1
    else:
        return pow(num,3) + sum(num-1)
a = int(input("enter a number:"))
result = sum(a)
print(f"{result} is the cube of natural numbers.")
```
