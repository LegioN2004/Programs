```python
def BinarySearch(arr, val):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (high + low) // 2
        if val == arr[mid]:
            return mid
        elif val < arr[mid]:
            high = mid - 1
        else:
            low = mid + 1
    return -1
    
a=[12,14,25,39,40,52,66]
num = int(input("Enter a number:"))
position = BinarySearch(a, num)
if position != -1:
    print("Number found at index:", position)
else:
    print("Number not found")
```
